```python
import numpy as np # linear algebra
import xml.etree.ElementTree as ET # for parsing XML
import matplotlib.pyplot as plt # to show images
from PIL import Image # to read images
import os
import glob
import cv2 as cv
```

```python
#from pathlib import Path
imgs="drive/MyDrive/dataset-master/JPEGImages"
annots="drive/MyDrive/dataset-master/Annotations"
files =  os.listdir(imgs)
simgs = sorted(files)
sanots = os.listdir(annots)
sannots = sorted(sanots)
print(simgs)
print(sannots)
def imageAnnotation():
  for ii in simgs:
    iii = ii
    imgg = os.path.splitext(iii)[0]
    for jj in sannots:
      jjj = jj
      xmll = os.path.splitext(jjj)[0]
      print('img =>',imgg)
      print('xml =>',xmll)
      if imgg == xmll:
        print('OKKKK',imgg)
        im = cv.imread(imgs+"/"+ii)
        plt.imshow(im)
        parser = ET.XMLParser(encoding="utf-8")
        targetTree = ET.parse(annots+"/"+jjj, parser=parser)
        rootTag = targetTree.getroot()
        ct = 0
        for obj in rootTag.iter('object'):
          xmin =obj.getchildren()[4][0].text
          ymin =obj.getchildren()[4][1].text
          xmax =obj.getchildren()[4][2].text
          ymax =obj.getchildren()[4][3].text
          crop = im[int(ymin):int(ymax), int(xmin):int(xmax)]
          print(ct)
          cv.imwrite(f'drive/MyDrive/dataset-master/annotated/'+str(imgg)+'_'+str({ct})+'_anotat.png',crop)
          ct = ct+1
          plt.imshow(crop)
        print('done')
    print('Annotation Done!!!')
```

```python
imageAnnotation()
```
